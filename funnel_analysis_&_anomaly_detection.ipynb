```python
import itertools

p_range = range(0, 6)
d_range = range(0, 6)
q_range = range(0, 6)

best_aic = float('inf')
best_pdq = None

for pdq in itertools.product(p_range, d_range, q_range):
    try:
        model = ARIMA(AP_data, order=pdq)
        arima_result = model.fit()
        aic = arima_result.aic
        if aic < best_aic:
            best_aic = aic
            best_pdq = pdq
    except:
        continue
print("Best AIC:", best_aic)
print("Best (p, d, q):", best_pdq)
```

#4..Clustering Analysis

```python
from sklearn.cluster import KMeans
```

```python
k_cols=['Actual Productivity','Overtime','No. of Workers']
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(funnel3[k_cols])
scaled_data
```

### Elbow Method

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
```

**By examining the WCSS values and the elbow plot, it appears that the optimal number of clusters is 3. This is because the rate of decrease in WCSS slows down significantly after the third cluster**

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
clusters
```

```python
k_df=funnel3[k_cols]
k_df['clusters']=clusters
k_df.head(3)
```

```python
k_df.clusters.value_counts()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(k_df['Actual Productivity'], k_df['Overtime'], c=k_df['clusters'], cmap='viridis', s=100)
plt.title('Clusters of Workers')
plt.xlabel('Actual Productivity')
plt.ylabel('Overtime')
plt.colorbar(label='Cluster')
plt.grid(True)
```

**The scatter plot shows how different clusters (colored points) are distributed based on Actual Productivity and Overtime**

**The plot of WCSS (Within-Cluster Sum of Squares) versus the number of clusters helps identify the optimal number of clusters. The "elbow" in the plot indicates the optimal number of clusters**

**Function to evaluate all metrics at once**

```python
y_pred2
```

```python
def evaluate_model(y_pred2, y_test1, model_name, metrics_df):
    # y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test1, y_pred2)
    precision = precision_score(y_test1, y_pred2)
    recall = recall_score(y_test1, y_pred2)
    f1 = f1_score(y_test1, y_pred2)
    # if hasattr(model, "predict_proba"):
    #     y_prob = model.predict_proba(X_test)[:, 1]
    roc_auc = roc_auc_score(y_test1, y_pred2)
    # else:
    #     roc_auc = None
    print('confusion_matrix\n',confusion_matrix(y_test1, y_pred2))
    print("-" * 30)
    new_row = pd.DataFrame({
        'Model': [model_name],
        'Accuracy': [accuracy],
        'Precision': [precision],
        'Recall': [recall],
        'F1 Score': [f1],
        'ROC-AUC': [roc_auc]
    })
    if new_row.Model.iloc[0] not  in metrics_df['Model'].values:
      metrics_df = pd.concat([metrics_df, new_row], ignore_index=True)
    print('classification_report\n',classification_report(y_test1, y_pred2, zero_division=1))
    print("-" * 30)
    return metrics_df
```

```python
metrics_df = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC-AUC'])
```

```python
metrics_df = evaluate_model(y_pred1, y_test, 'Isolation Forest', metrics_df)
metrics_df
```

## Autoencoders (for deep learning)

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers
```

```python
input_dim = X_train.shape[1]
encoding_dim = 14
```

```python
input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu', activity_regularizer=regularizers.l1(10e-5))(input_layer)
decoded = Dense(input_dim, activation='sigmoid')(encoded)
autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
autoencoder.fit(X_train, X_train, epochs=50, batch_size=256, shuffle=True, validation_data=(X_test, X_test))
```

```python
X_train_pred = autoencoder.predict(X_train)
X_test_pred = autoencoder.predict(X_test)
X_test_pred
```

```python
train_loss = np.mean(np.square(X_train - X_train_pred), axis=1)
test_loss = np.mean(np.square(X_test - X_test_pred), axis=1)
display(train_loss.head(3))
test_loss.head(3)
```

```python
threshold=np.percentile(train_loss, 99)
threshold
```

```python
y_pred_test=[1 if x > threshold else 0 for x in test_loss]
pd.DataFrame(y_pred_test).value_counts()
```

```python
y_test.value_counts()
```

```python
metrics_df = evaluate_model(y_pred_test, y_test.values, 'Autoencoders', metrics_df)
metrics_df
```

* **Balanced metrics with a reasonable recall (0.531) and decent ROC-AUC (0.761). The precision (0.089) and F1 Score (0.153) are also better than the other models**

**In Isolation Forest**
 * **Very low precision (0.034) and F1 Score (0.066). The low precision means many of its fraud predictions are incorrect, which could lead to a high number of false positives**
